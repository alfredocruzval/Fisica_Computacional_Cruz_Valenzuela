# Lagrange

```julia
using SymPy
```

```julia
x=Sym("x") #Sea x una variable simbólica
function Lagrange(list1,list2,x) #definimos función con la que se trabajará
array1=[]
array2=[]  #Arreglos que se utilizarán
array3=[]
Polinomio=0
tot=0
   
for a in 1:length(list1) # ciclo para la multiplicación de Lagrange
tot2=1 # variable del total para reiniciación
for b in 1:length(list2) # ciclo a recorrer con j
if b!=a # si b es distinto de a entonces calculará la base
Base=(-list1[b]+x)/ (list1[a]-list1[b]) # Con esta formula se calculará la base
tot2=Base*tot2 # aquí se acumula para que posteriormente se multipliquen las bases
            end
        end
        
    push!(array1,tot2) 
    end
    
    for i in 1:length(list2)
        
Polinomio= array1[i]*list2[i] #Formulas para polinomio y el total
tot= Polinomio+tot
        
    end
    
    tot=simplify(tot)
    
    return tot #Se obtiene tot    
end
```

```julia
list1=[0,2,3,4,5,6,7]
list2=[10,-10,-4,0,5,20,-1] #Con los valores que nos pide
Lagrange(list1,list2,x)
```

```julia
using PyPlot #Gráfica del código
Y=simplify(Lagrange(list1,list2,x))
X=linspace(0,7,100)
Y2=Y.(X)

plot(X,Y2)
scatter(list1,list2,color="red")
grid()
```
