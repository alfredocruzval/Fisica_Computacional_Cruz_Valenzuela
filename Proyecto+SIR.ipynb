## Introducción ##

### Modelo simple de transmisión en una población ###

Sea el caso en el cual existe algún agente infeccioso que se peude transmitir a cualquier individuo susceptible. Entonces se define para cada tiempo t, se devide a la población en tres clases epidemiológicas: los individuos susceptibles de infección (S), los infectados (I), y aquellos individuos que han dejado de tener la infección (R), por muerte o por adquisición de inmunidad permanente.

Sea la suma de las tres subpoblaciones igual a N, tal que N es el número total de individuos (N = S + I + R). Suponiendo que un cierto número de individuos susceptibles se infectan en un tiempo h a una tasa $\alpha$, entonces el número de individuos en el tiempo t + h está dado por

$$ S(t+h) = S(t)(1-\alpha h)$$

donde el número de nuevos individuos infectados entre t y t + h es entonces h$\alpha$S(t). Si suponemos que en ese mismo periodo de tiempo, una proporción $\delta$
de los infectados deja de serlo, el número de individuos infectados cambia a:

$$I(t+h)=h\alpha S(t) + (1-h\delta)I(t)$$

Por otro lado el cambio correspondiente en el número de individuos que ya no tienen la infección es:

$$R(t+h)=R(t)+h\delta I(t)$$

#### Fuerza de infección y exposición homogénea.####
El término $\alpha$ en la ecuación se le conoce como fuerza de infección, y esta puede depender de varios factores. Una suposición aplicable a varias enfermedades infecciosas
es que  $\alpha$ depende de la proporción de individuos infectados en la población, de la tasa de contacto con el patógeno que causa la enfermedad, y la probabilidad de infección dada una exposición directa con el patógeno. Entonces, suponiendo que los individuos que dejan de tener la infección adquiereninmunidad permanente

$$\alpha=\beta \frac{I(t)}{S(t)+I(t)+R(t)}$$

En contraste, si la infección no tiene curación y es mortal, la fuerza de infección se puede escribir como:

$$\alpha=\frac{\beta I(t)}{S(t)+I(t)}$$

la última ecuación se puede interpretar como el producto
de la probabilidad de tener contacto con un individuo infectado multiplicada por la probabilidad de infección dado el contacto con el patógeno, o el individuo que lo porta, representada por $\beta$.

Entonces suponemos que cualquier persona susceptible tiene la misma probabilidad de encontrarse con un infectado, y viceversa.

#### Dinamica de transmisión y densidad poblacional ####

Si se supone que las poblaciones son lo suficientemente grandes, entonces es posible substituir S, I, y R por densidades poblacionales u, v, y w, respectivamente.

Si N es lo suficientemente grande y constante, es posible dividir las primeras tres ecuaciones entre N,
combinar la fuerza de infección $\alpha$, y calcular el límite cuando h tiende a cero para obtener ecuaciones diferenciales ordinarias definidas como:

 $$\frac{du}{dt}= -\beta vu$$
 $$\frac{dv}{dt}= (\beta u - \delta) v$$
 $$\frac{dw}{dt}= \delta v$$
 
#### Brotes y epidemias ####

Se define brote como un conjunto de individuos infectados que no crece.

Se define epidemia sólo cuando la densidad de población infectada aumenta.

Un brote entonces representa condiciones iniciales para el sistema en las que v > 0. Una epidemia ocurre, en consecuencia, cuando $\frac{dv}{dt} > 0$, que pasa cuando $\beta$u − $\delta$ > 0. 

En cambio, una epidemia no puede ocurrir cuando $\frac{dv}{dt}$ menor-igual a 0, aunque haya un brote epidémico.

## Modelo SIR

En este primer modelo, se tomará en cuenta a la enfermedad conocida como Influenza, tomando como variable independiente al tiempo. Tomando como conjunto de variables dependientes cuenta personas en cada uno de los grupos, cada una como función del tiempo:

S = S(t) es el número de individuos susceptibles
I = I(t) es el número de individuos infectados
R = R(t) es el número de individuos recuperados

![Ilustracion modelo SIR](1.png) Modelo SIR simple. Individuos susceptibles S, infectados y recuperados. El parámetro B representa el valor per cápita por tasa infecciosa gamma denota la recuperación per cápita

Por otro lado, estableciendo unas cuantas suposiciones veremos como vambian las tasas de cambio de nuestras variables independientes

Nadie se agrega al grupo susceptible, ya que estamos ignorando los nacimientos y la inmigración. La única forma en que un individuo abandona el grupo susceptible es al infectarse.

Suponemos que la tasa de cambio de tiempo de S (t) , el número de susceptibles, 1 depende del número ya susceptible, la cantidad de individuos ya infectados y la cantidad de contacto entre personas susceptibles e infectadas. En particular, supongamos que cada individuo infectado tiene un número fijo de contactos b por día que son suficientes para propagar la enfermedad.

No todos estos contactos son con individuos susceptibles. Si asumimos una mezcla homogénea de la población, la fracción de estos contactos que están con susceptibles es s (t) . Por lo tanto, en promedio, cada individuo infectado genera b s (t) nuevos individuos infectados por día. 

Observación: Con una gran población susceptible y una población infectada relativamente pequeña, podmos ignorar situaciones complicadas de conteo, como un solo individuo susceptible que se encuntre con más de un infectado en un día determiando.

También suponemos que una fracción fija k del grupo infectado se recuperará durante un día determinado. Por ejemplo, si la duración promedio de la infección es de tres días, entonces, en promedio, un tercio de la población actualmente infectada se recupera cada día. 


Finalmente, completamos nuestro modelo asignando a cada ecuación diferencial una condición inicial. 

Para este caso en particular -una epidemia de gripe en alguna ciudad, casi nadie era inmune al comienzo de la epidemia, por lo que casi todos eran susceptibles. 

Supondremos que hubo un rastro de nivel de infección en la población, por ejemplo, 10 personas. Por lo tanto, nuestros valores iniciales para las variables de población son

S(0) = 7,900000

I(0) = 10

R(0) = 0

En términos de las variables escaladas, estas condiciones iniciales son:

S(0) = 1       

I(0) = 1.27x10^-6 

R(0) = 0$

 $$\frac{ds}{dt}= -bs(t)i(t)$$
 $$\frac{di}{dt}= bs(t)i(t)-ki(t)$$
 $$\frac{dr}{dt}=  ki(t)$$
 
El periodo promedio de infecciosidad a los tres días, es k = 1/3 . Si suponemos que cada infectado haría un posible contacto infectante cada dos días, entonces b sería 1/2 . Hacemos hincapié en que esto es solo una suposición.


Usaremos esta función para pasar los valores a y k a las funciones de actualización. Es decir, al escribir el código para la función, solo le pasamos la argumentación S, I, R y tomamos los valores de los parámetros del contexto en el que se llama a la función. Esto es sorprendentemente conveniente, especialmente cuando queremos experimentar con diferentes valores de parámetros con el fin de ajustar la salida del modelo a los datos

```julia
using PyPlot
```

```julia
function SIR(condicionesiniciales) #Se declara una función
    s=condicionesiniciales[1];
    i=condicionesiniciales[2];
    r=condicionesiniciales[3];
    n=s+i+r
    NS = s+(-b*s*i*dt)
    NI = i+ (b*s*i-k*i*dt)
    NR = r+ (k*i*dt)
    return [NS,NI,NR]
end
```

```julia
b = 1/2
k = 1/3
dt = 0.5
tfinal = 100
i0 = 0.00000127
s0 = 1-i0
r0 = 0

nsteps = round(Int64, tfinal/dt)
resultvals = Array(Float64, nsteps+1, 3)
timevec = Array(Float64, nsteps+1)
resultvals[1,:] = [s0,i0,r0]
timevec[1] = 0

for step = 1:nsteps
    resultvals[step+1, :]= SIR(resultvals[step, :])
    timevec[step+1] = timevec[step] + dt
end
```

```julia
title("SIR")
xlabel("Tiempo(días)")
ylabel("Población")
plot(timevec,resultvals[:,1],label= "Susceptibles")
plot(timevec,resultvals[:,2],label= "Infectados")
plot(timevec,resultvals[:,3],label= "Recuperados")
legend()
grid()
show()
```

### Modelo SIS

Esta es una simplificación del modelo $SIR$ con la diferencia de que los individuos nunca pueden generan inmuninad a la enfermedad, un ejemplo es la gripe común, a su vez es te modelo contempla el nacimiento y la muerte de las personas, el flujo de poblacón se ve de la siguiente forma:$$S->I->S$$ El sistema de ecuaciones es de la siguiente forma:

$$\frac{dS}{dt} = -\beta SI + \mu(N-S)+\gamma I$$

$$\frac{dI}{dt} = \beta SI-\gamma I-\mu I$$

```julia
function SIS(popnvector)#Se genero una función para solucionar el sistema de ecuaciones
    s=popnvector[1];#susceptibles
    i=popnvector[2];#infectados
    n=s+i
    NewS = s - beta*s*i*dt + mu*(n-s)*dt + gam*i*dt
    NewI = i+ beta*s*i*dt - gam*i*dt -mu*i*dt
    return [NewS, NewI]
end
```

```julia
beta = 1/20000 #TProbabilidad de contagiarse
gam = 1/10#Tiempo de infección
dt = 0.01#Pasos en el tiempo
mu = 0.034#Tasa de mortalidad por infección
tfinal = 70#Condición de alto
#Condiciones iniciales
s0 = 0.7
i0 = 0.3


nsteps = round(Int64, tfinal/dt)
resultvals = Array(Float64, nsteps+1, 2)
timevec = Array(Float64, nsteps+1)
resultvals[1,:] = [s0,i0]
timevec[1] = 0

for step = 1:nsteps
    resultvals[step+1, :]= SIS(resultvals[step, :])
    timevec[step+1] = timevec[step] + dt
end
```

```julia
title("SIS")
xlabel("Tiempo(días)")
ylabel("Población")
plot(timevec,resultvals[:,1],label= "Susceptibles")
plot(timevec,resultvals[:,2],label= "Infectados")
legend()
grid()
show()
```

### Modelo SIRS

En lugar de ser una simplificación como el modelo anterior esta es una extensión y la dinámica de este modelo es $$S->I->R->S$$ El sistema de ecuaciones para un modelo $SIRS$ es:

$$\frac{dS}{dt} = -\beta SI + fR$$

$$\frac{dI}{dt} = \beta SI-\gamma I-\mu I$$

$$\frac{dR}{dt} =\gamma I-\mu R-fR$$

```julia
function SIRS(popnvector)#Se genero una función para solucionar el sistema de ecuaciones
    s=popnvector[1];#susceptibles
    i=popnvector[2];#infectados
    r=popnvector[3];#infectados
    n=s+i+r
    NewS = s - beta*s*i*dt  + f*r*dt
    NewI = i+ beta*s*i*dt - gam*i*dt -mu*i*dt
    NewR = r+ gam*i*dt -mu*r*dt -f*r*dt
    return [NewS, NewI, NewR]
end
```

```julia
beta = 1/20000 #TProbabilidad de contagiarse
gam = 1/10#Tiempo de infección
dt = 0.01#Pasos en el tiempo
mu = 0.034#Tasa de mortalidad por infección
f = 1/30#Tasa de perdida de inmunidad
tfinal = 70#Condición de alto
#Condiciones iniciales
s0 = 0.7
i0 = 0.3
r0 = 0

nsteps = round(Int64, tfinal/dt)
resultvals = Array(Float64, nsteps+1, 3)
timevec = Array(Float64, nsteps+1)
resultvals[1,:] = [s0,i0,r0]
timevec[1] = 0

for step = 1:nsteps
    resultvals[step+1, :]= SIRS(resultvals[step, :])
    timevec[step+1] = timevec[step] + dt
end
```

```julia
title("SIRS")
xlabel("Tiempo(días)")
ylabel("Población")
plot(timevec,resultvals[:,1],label= "Susceptibles")
plot(timevec,resultvals[:,2],label= "Infectados")
plot(timevec,resultvals[:,3],label= "Recuperados")
legend()
grid()
show()
```

### Modelo SEIR

El modelo $SEIR$ el cual es una extensión del modelo $SIR$ con una nueva población $E$ que serán el número de personas expuestas al virus. Este modelo se comporta de la siguiente dinámica: $$S->E->I->R$$ El sistema de ecuaciones de este modelo se ve de la siguiente forma:
$$\frac{dS}{dt} = - \frac{\beta S I}{N}$$

$$\frac{dI}{dt} = \delta E- \gamma I$$

$$\frac {dE}{dt} = \frac{\beta S I}{N}-\delta E$$

$$\frac{dR}{dt} = \gamma I$$

Esto donde $\beta$ es la probabilidad de contagio, $\frac{1}{\delta}$ es el periodo de incubación y $\frac{1}{\gamma}$ es el periodo de infección.

Algunas de las consideraciones más importantes es que para este modelo, son que la población $R$ significa removidos, es decir que no pueden regresar a ser susceptibles y que el tamaño de la población es constante, por lo tanto $$N(t)=S(t)+E(t)+I(t)+R(t)=cte$$



### Modelo SEIR aplicado a la modelación del Ebola

En este trabajo se modelará el comportamiento de una epidemia por EVE.En el 2014 en el oeste de África tuvo lugar una de las mayores epidemias de Ebola(EVE) en la historia, se utilizarón datos de Sierra Leona y Liberia para la parametrización del modelo.

```julia
using PyPlot
function SEIR(condicionesiniciales)#Se genero una función para resolver el sistema de ecuaciones del modelo SERI
    s=condicionesiniciales[1];#Se declaran las condiciones iniciales de cada población para fines practicos se tomaran numeros del 0 al 6 pero es importante recordar que son millones de personas
    i=condicionesiniciales[2];
    e=condicionesiniciales[3];
    r=condicionesiniciales[4];
    n = s+i+r+e
    NewS =s  - (beta*s*i)/n*dt#Suceptibles
    NewI = i + d*e*dt - gam*i*dt#Infectados
    NewE = e + (beta*s*i)/n*dt - d*e*dt#Expuestos
    NewR = r + gam*i*dt#Removidos
#dt representa el cambio en el tiempo en días
    return[NewS, NewI, NewE, NewR]
end
```

```julia
#Parametrización del modelo en Sierra Leona
dt = 0.5
beta =0.128#Probabilidad de contagio
gam = 1/10.38 #Periodo de infección (1/gam)
d =1/10 #Periodo de incubación(1/d)
tfinal = 250#Es la condicón de alto, en esta ocasión será de 250 días
s0 = 5.4828#millones de personas
i0 = 0#millones de personas
e0 = 0.6092#millones de personas
r0=0#millones de personas


nsteps = round(Int64, tfinal/dt)
resultvalsSL = Array(Float64, nsteps+1, 4)
timevec = Array(Float64, nsteps+1)
resultvalsSL[1,:] = [s0,i0,e0,r0]
timevec[1] = 0

for step = 1:nsteps
    resultvalsSL[step+1, :]= SEIR(resultvalsSL[step, :])
    timevec[step+1] = timevec[step] + dt
end
```

```julia
title("Comportamiento del Ebola Sierra Leona")
plot(timevec,resultvalsSL[:,1],label="Susceptibles")
plot(timevec,resultvalsSL[:,2],label="Infectados")
plot(timevec,resultvalsSL[:,3],label="Expuestos")
plot(timevec,resultvalsSL[:,4],label="Removidos")
xlabel("Tiempo(días)")
ylabel("Población")
grid()
legend()
show()
```

En la gráfica anterior podemos ver que la cantidad de fallecidos en ambas regiones es extremadamente grande y los recuperados son muy pocos.Por esta razón es que se pueden tomar medidas como la cuarntena, esto quiere decir que se podrían separar a la mayor cantidad de personas infectadas para evitar que continue el contagio, lo que ayudaría a evitar una epidemia.

Podemos anexar una nueva variable  $Q$ al modelo, la cual representa a las personas hospitalizadas por el gobierno o alguna otra organización, y una cantidad $\alpha$ la cual representa la proporción de infectados hozpitalizados por lo que el modelo queda de la siguiente forma:

$$\frac{dS}{dt} = - \frac{\beta}{\gamma}S(1-\alpha)I$$

$$\frac{dI}{dt} = \frac{\delta}{\gamma}E-I$$

$$\frac {dE}{dt} = \frac{\beta}{\gamma}S(1-\alpha)I-\frac{\delta}{\gamma}E$$

$$\frac{dR}{dt} = I$$

$$ \frac{dQ}{dt} = \alpha I$$


```julia
function SIREQ(popnvector)
    s=popnvector[1];
    i=popnvector[2];
    e=popnvector[3];
    r=popnvector[4];
    q=popnvector[5];
    n = s+i+r+e
    NewS =s  - ((beta/gam)*s*(1-alfa)*i)*dt
    NewI = i + (d/gam)*e*dt - i*dt
    NewE = e + ((beta/gam)*s*(1-alfa)*i)*dt - (d/gam)*e*dt
    NewR = r + i*dt
    NewQ = q + alfa*i*dt

    return[NewS, NewI, NewE, NewR,NewQ]
end
```

```julia
#Datos Sierra Leona
dt = 0.05
beta =0.128
gam = 1/10.38
d =1/10
alfa = 0.7
tfinal = 20
s0 = 5.4828
i0 = 0
e0 = 0.6092
r0=0
q0=0

nsteps = round(Int64, tfinal/dt)
resultvalsSL = Array(Float64, nsteps+1, 5)
timevec = Array(Float64, nsteps+1)
resultvalsSL[1,:] = [s0,i0,e0,r0,q0]
timevec[1] = 0

for step = 1:nsteps
    resultvalsSL[step+1, :]= SIREQ(resultvalsSL[step, :])
    timevec[step+1] = timevec[step] + dt
end
```

```julia
title("Comportamiento del Ebola Sierra Leona con cuarentena")
plot(timevec,resultvalsSL[:,1],label="Susceptibles")
plot(timevec,resultvalsSL[:,2],label="Infectados")
plot(timevec,resultvalsSL[:,3],label="Expuestos")
plot(timevec,resultvalsSL[:,4],label="Removidos")
plot(timevec,resultvalsSL[:,5],label="Hospitalizados")
xlabel("Tiempo(días)")
ylabel("Población")
grid()
legend()
show()
```

Es importante destacar que no importa que tan pequeña sea $\alpha$ la epidemia siempre se controla en un intervalo de tiempo muy corto.

### Modelo MESIR

Esta derivación del modelo $SEIR$ contempla una nueva población $M$, la cual considera a los recien nacidos con inmunidad por un periodo de tiempo $\delta$, el flujo de este modelo se ve de la siguiente forma: $$M->S->E->I->R$$

Y las ecuaciones diferenciales:

$$\dot M = B -\delta MS -\mu M$$

$$\dot S = \delta MS -\beta SI - \mu s$$

$$\dot E = \beta SI -(\epsilon+\mu)E$$

$$\dot I = \epsilon E - (\gamma +\mu)I$$

$$\dot R = \gamma I - \mu R $$

```julia
function MSEIR(condicionesiniciales)#Se genero una función para resolver el sistema de ecuaciones del modelo SERI
    s=condicionesiniciales[1];#Se declaran las condiciones iniciales de cada población para fines practicos se tomaran numeros del 0 al 6 pero es importante recordar que son millones de personas
    i=condicionesiniciales[2];
    e=condicionesiniciales[3];
    r=condicionesiniciales[4];
    m=condicionesiniciales[5];
    n = s+i+r+e
    NewS =s + delta*m*s*dt - (beta*s*i)*dt - mu*s+dt#Suceptibles
    NewI = i + epsilon*e*dt - (gam+mu)*i*dt#Infectados
    NewE = e + (beta*s*i)*dt - (epsilon+mu)*e*dt#Expuestos
    NewR = r + gam*i*dt - mu*r*dt#Removidos
    NewM = m + B - delta*m*s*dt - mu*m*dt
#dt representa el cambio en el tiempo en días
    return[NewS, NewI, NewE, NewR, NewM]
end
```

Para realizar la muestra de este modelo se supuso que los bebes puedes ser inmunes al ébola por lo menos algunos días después de su nacimiento, se recopilaron datos del promedio para conocer la tasa de natalidad en Sierra Leona , lo que se encontró es que a lo largo del 2014 se tuvo una tasa de 4.63 hijos por mujer y considerando que en esta ciudad de cada hay 4.85 por cada mujer entonces en total hay 1.256082 mujeres en Sierra Leona, por lo tanto, $B=0.00159333201$ por día.

```julia
#Parametrización del modelo en Sierra Leona
dt = 0.5
beta =0.128#Probabilidad de contagio
gam = 1/10.38 #Periodo de infección (1/gam)
epsilon = 1/10#periodo de incubación(1/epsilon)
delta = 0.001 #Periodo de inmunidad temporal
B = 0.00159333201 #tasa promedio de nacimientos
mu =0.85 #tasa de mortalidad por la enfermedad
tfinal = 250#Es la condicón de alto, en esta ocación será de 250 días
s0 = 5.4828#millones de personas
i0 = 0#millones de personas
e0 = 0.6092#millones de personas
r0=0#millones de personas
m0=0#millones de personas

nsteps = round(Int64, tfinal/dt)
resultvalsSL = Array(Float64, nsteps+1, 5)
timevec = Array(Float64, nsteps+1)
resultvalsSL[1,:] = [s0,i0,e0,r0,m0]
timevec[1] = 0

for step = 1:nsteps
    resultvalsSL[step+1, :]= MSEIR(resultvalsSL[step, :])
    timevec[step+1] = timevec[step] + dt
end
```

```julia
title("Modelo MSEIR")
plot(timevec,resultvalsSL[:,1],label="Susceptibles")
plot(timevec,resultvalsSL[:,2],label="Infectados")
plot(timevec,resultvalsSL[:,3],label="Expuestos")
plot(timevec,resultvalsSL[:,4],label="Removidos")
plot(timevec,resultvalsSL[:,5],label="Nacidos con inmunidad")
xlabel("Tiempo(días)")
ylabel("Población")
grid()
legend()
show()
```

### Modificación del Modelo SIR para una epidemia por dengue.

Este modelo es una modificación del modelo $SIR$ lo novedoso de este modelo creado por investigadores de la universidad de Arizona y de Costa Rica, es que contempla a un vector de infección (el mosquito $Ae. Aegypti$), dos posibles enfermedades la DENV-2 tanto en su genotipo americano como el asiático, tambien considera que las personas pueden recuperarse o progresar a DHF (Dengue hemorrhagic fever), pero lo que sin duda hace que este modelo sea distinto a todos los que se habían hecho hasta el momento es que considera una transmisión vertical. A lo que se refiere es lo siguiente: 

En general los modelos $SIR$ contemplan un vector de infección o agente infeccioso, en este caso el mosquito, este agente es el que es capaz de infectar a las personas, transmisión horizontal (mosquito->persona) tambien considera en sentido opuesto ya que si un mosquito pica a una persona infectada este también se infecta(persona ->mosquito) . Pero que pasa si además de infectar a las personas este agente puede pasar el virus a la siguiente generación, la transmisión vertica (mosquito->mosquito). Esto aumenta considerablemente la probabilidad de contraer DENV-2. El sistema de ecuaciónes diferenciales queda de la siguiente forma:

$$\dot S = \mu N - \frac{\beta_{Am}SW_{Am}}{M}-\frac{\beta_{As}SW_{As}}{M}-\mu S$$

$$\dot D_{Am} =\frac{\beta_{Am}SW_{Am}}{M}-(\delta_{Am}+\mu) D_{Am}$$

$$\dot D_{As} =\frac{\beta_{As}SW_{As}}{M}-(\delta_{As}+\alpha+\mu) D_{As}$$

$$\dot H = \alpha D_{Am} - (\delta_H + \mu)H$$

$$\dot R = \delta_{Am}D_{Am} + \delta_{As}D_{As} +\delta_HH-\mu R$$

$$\dot V = \mu_mM - p\mu_mW_{Am}-\frac{V\theta_{Am}D_{Am}}{N}-\frac{V\theta_{As}D_{As}}{N}-\mu_mV$$

$$\dot W_{Am} =\frac{V\theta_{Am}D_{Am}}{N}-\mu_{m}W_{Am}$$

$$\dot W_{As} =\frac{V\theta_{As}D_{As}}{N}+p\mu_mW_{As}-\mu_{m}W_{As}$$

Para la parametrización de este modelo se tomaron los datos de la epidemia de dengue en Perú en el año 2000.

![Ilustración Modelo huésped SIR DENGUE](sir.png) Modelo del huésped Susceptible S, infectado con DENV 2 genotipo estadounidense, DAm, infectado con DENV-2 genotipo asiático, DAs progresión a DH y recuperado, R, individuos.

![Ilustración Modelo vectorial SIR DENGUE](sir2.png) Modelo de vector Susceptible V, infectado con el genotipo estadounidense DENV 2, WAm, infectado con el genotipo asiático DENV-2, WAs. Una cierta proporción p, o nacimientos, confiere transmisión vertical y entra en WA directamente.


```julia
function SIRDengue(popnvector)#Creamos una función para solucionar el sistema de ecuaciones.
    M = 1#Población total de mosquitos(como es muy complicado saberla para este modelo trabajaremos con proporciones) 
    N = 1 #Población total de personas
    s=popnvector[1];#Susceptibles
    dam=popnvector[2];#Personas infectadas con DENV-2 americano
    das=popnvector[3];#Personas infectadas con DENV-2 asático
    h=popnvector[4];#Personas que progresan a DHF
    r=popnvector[5];#Recuperados
    v=popnvector[6];#Mosquitos susceptibles
    wam=popnvector[7];#Mosquitos infectadas con DENV-2 americano
    was=popnvector[8];#Mosquitos infectadas con DENV-2 asiatico
    #Sistema de ecuaciones.
    NewS =s + (mu*N-(betam*s*wam)/M-(betas*s*was)/M-mu*s)*dt
    NewDam = dam + ((betam*s*wam)/M-(deltam+mu)*dam)*dt
    NewDas = das + ((betas*s*was)/M-(deltas+mu+alfa)*das)*dt
    NewH = h + (alfa*das-(deltah+mu)*h)*dt
    NewR = r +(deltam*dam+deltas*das+deltah*h-mu*r)*dt
    NewV = v + (mum*M-p*mum*wam-(v*thetam*dam)/N-(v*thetas*das)/N-mum*v)*dt
    NewWam = wam +((v*thetam*dam)/N-mum*wam)*dt
    NewWas = was +((v*thetam*dam)/N + p*mum*was-mum*was)*dt

    return[NewS, NewDam, NewDas, NewH, NewR, NewV, NewWam, NewWas]
end
```

```julia
dt = 0.5#Cambios en el tiempo
mu =0.0048#Tasa de Natalidad/Mortalidad natural por dia en Humanos
thetas = 0.8160#Tasa de infeccion persona -> mosquito del genotipo asiatico por día
thetam = 1.6212#Tasa de infeccion persona -> mosquito del genotipo americano por día
betas = 0.0483#Tasa de infeccion persona <- mosquito del genotipo asiático por día
betam = 0.0182#Tasa de infeccion persona <- mosquito del genotipo americano por día
alfa = 0.113#Tasa de progresión del genotipo asiatico a DHF por día
mum = 0.0958#Tasa de mortalidad del mosquito por día
p = 0.103#Proporción de mosquitos infectados por transmisión vertical
deltam=0.2#Tasa de recuperación del genotipo americano por día
deltas=0.2#Tasa de recuperación del genotipo asiático por día
deltah=0.2#Tasa de recuperación de DHF por día
tfinal = 250
#Condiciones iniciales
s0 = 0.9
dam0 = 0.05
das0 = 0.05
h0=0
r0=0
v0=0.9
was0=0.05
wam0=0.05


nsteps = round(Int64, tfinal/dt)
resultvals = Array(Float64, nsteps+1, 8)
timevec = Array(Float64, nsteps+1)
resultvals[1,:] = [s0,dam0,das0,h0,r0,v0,wam0,was0]
timevec[1] = 0

for step = 1:nsteps
    resultvals[step+1, :]= SIRDengue(resultvals[step, :])
    timevec[step+1] = timevec[step] + dt
end
```

```julia
title("Dinámica de las Personas")
plot(timevec,resultvals[:,1],label="Susceptibles")
plot(timevec,resultvals[:,2],label="Infectados con Am")
plot(timevec,resultvals[:,3],label="Infectados con As")
plot(timevec,resultvals[:,4],label="Personas con DHF")
plot(timevec,resultvals[:,5],label="Recuperados")
xlabel("Tiempo(días)")
ylabel("Población")
grid()
legend()
show()
```

```julia
title("Dinámica de los Mosquitos")
plot(timevec,resultvals[:,6],label="Susceptibles")
plot(timevec,resultvals[:,7],label="Infectados con Am")
plot(timevec,resultvals[:,8],label="Infectados con As")
xlabel("Tiempo(días)")
ylabel("Población")
grid()
legend()
show()
```
