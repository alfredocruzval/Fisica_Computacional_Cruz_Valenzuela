```julia
using Plots #Se uso plots porque quedan mejor las gráficas que con pyPlot
```

```julia
function metodo_euler(f,xo,to,tf,h) #Condiciones función
    ##listt = []
    listx = [xo]
    listt=to:h:tf #Arreglos
    for i in 1:length(listt)-1 #Ciclo con el que trabajará el código
    xo=xo+h*f((xo),listt[i])  #Euler
        push!(listx,xo)
    end
    return listt,listx #Se obtiene lo que se quiere
end    
```

```julia
f(x,t)=-2.5x #alpha #función a evaluar
```

```julia
h0=metodo_euler(f,1.0,0.0,8.0,1.0) #puntos condición inicial
h1=metodo_euler(f,1.0,0.0,8.0,0.5)
h2=metodo_euler(f,1.0,0.0,8.0,0.2)
h3=metodo_euler(f,1.0,0.0,8.0,0.01)
hE(t)=e^(-2.5*t)  #Solución
```

```julia
plot(h0,label="h=1.0")
plot!(linspace(0,8,100),hE)
#Se grafican los puntos y se compararán posteriormente
```

```julia
plot(h1,label="h=1.0")
plot!(linspace(0,8,100),hE)
```

```julia
plot(h2,label="h=1.0")
plot!(linspace(0,8,100),hE)
```

```julia
plot(linspace(0,8,50),hE)
plot!(h1,label="h=0.5")
plot!(h2,label="h=0.2")
plot!(h3,label="h=0.01")
```

Ejercicio 3

Haz $u = \dot{x}(t)$ y calcula la derivada de $u$ ¿qué obtienes y como se relaciona con el sistema original?


Utilizando que $u=\dot{x}(t)$
calculando la derivada se obtiene lo siguiente:
$$\dot{u}=\ddot{x}=-kx $$

con esta igualdad se llega a un sistema de tal forma que la primera derivada de u es igual a -kx. Mientras que la derivada de x es igual a u. Así como se ve a continuación:

$$\dot{x}=u$$
$$\dot{u}=-kx$$ 


Ejercicio 4

Obten la solución del sistema $\ddot{x} = -kx(t)$ utilizando lo escrito arriba y el método de Euler.

Para utilizar el método de Euler, impleméntelo para sistemas de más de una dimensión.

```julia
f(x,t)=-x #Código implementando ejercicio 3 con método de Euler
g(t,u)=u
function metodo_euler2(f,g,xo,uo,to,t,h)
    listt=to:h:t
    list1=[xo] #Arreglos
    list2=[uo]
    for i in 1:length(listt)-1
    uo=uo+h*f((xo),listt[i])  #Euler para cada una de las
    xo=xo+h*g((uo),listt[i])  #funciones
        push!(list1,xo)
        push!(list2,uo)
    end
    return listt,list1  #lo que se quiere obtener
end
```

```julia
h0=metodo_euler2(f,g,0.0,1.0,0.0,pi,1) #puntos a graficar
h1=metodo_euler2(f,g,0.0,1.0,0.0,pi,0.5)
h2=metodo_euler2(f,g,0.0,1.0,0.0,pi,0.1)
hs(t)=cos(t)
```

```julia
plot(linspace(0,pi,200),hs)
plot!(h0,label="h=1")   
plot!(h1,label="h=0.5")
plot!(h2,label="h=0.1")
```
