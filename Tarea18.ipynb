```julia
function bifurcacion(x,r,n)                    #Funci贸n bifurcaci贸n
    for i in 1:n                              
        x = r*x*(1-x)                          #Esta es la curva logistica
    end
    return x                                   #Lo que se quiere obtener
    end
```

```julia
function Lyapunov(x,n,r)                      #Funci贸n Lyapunov
    l=0    
for i in 1:n                                 
    x = r*x*(1-x)                             #Curva logistica
    l = l+log(abs(r*(1-2*x)))                 #Exponente de Lyaponuv
    end
    return l/n                              #Lo que se quiere obtener
    end
```

```julia
A=linspace(0,1,1000)
B=linspace(1,4,1000)
list1 =[]
list2 =[]
for i in 1:length(B)
    push!(list1,bifurcacion.(A,B[i],1000))
    push!(list2,ones(length(A))*B[i])
    end

#Iteraciones para poder graficar.
```

```julia
using PyPlot
plot(list2,list1,"o",markersize=0.05)
plot(B,Lyapunov.(0.5,1000,B),"o",markersize=0.5)
title("Bifurcaci贸n")
title("Lyapunov")
grid()
```
