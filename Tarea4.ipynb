```julia
function     F(x)
    return x^3-x^2-x-1 #Se declaran la función y la derivada de la misma función
end
```

```julia
function     dF(x)
    return 3x^2-2x-1
end
```

```julia
function Newt(x0,N) #En esta parte se calculan las raíces
    
    F(x)=x^3-x^2-x-1
    dF(x)=3x^2-2x-1
    list=zeros(N)  #Arreglo
        for i in 1:N #Ciclo for en el que trabajará el programa
            x0=x0-(F(x0)/dF(x0)) #método de Newton
        list[i]=x0
        end
    return list #Se obtiene la variable list
end
```

```julia
Newt(5,10)  #Se prueba el código ingresando condiciones iniciales
```

```julia
list1=Newt(5,84); #Se generan cuatro arreglos con las cndiciones
list2=Newt(23,87);
list3=Newt(12,44);
list4=Newt(-11,35);
```

```julia
using PyPlot
```

```julia
plot(real(list1)) # Se grafican los arreglos ingresados anteriormente
plot(real(list2))
plot(real(list3))
plot(real(list4))
show();
```

```julia
function NewtEr(x0,N) #En esta parte se calculan las raíces
    F(x)=x^3-x^2-x-1
    dF(x)=3x^2-2x-1
    list=zeros(N)
        for i in 1:N
            x0=x0-(F(x0)/dF(x0))
        list[i]=x0
        end
    return (list-list[N])./list[N] #Se obtiene el error
end
```

```julia
NewtEr(8,12) #Se prueba el código
```

```julia
list1=NewtEr(15,44); #Se generan los arreglos para la parte del error
list2=NewtEr(19,52);
list3=NewtEr(12,66);
list4=NewtEr(-23,49);
```

```julia
plot(real(list1)) # Se grafican los arreglos del error
plot(real(list2))
plot(real(list3))
plot(real(list4))
show();
```

# Bisección

```julia
function biseccionraices(a,b,N)#Se obtienen las raíces por bisección
list= [] 
p = (a+b)/2
    f(x) = x^3-x^2-x-1
    for i in 1:N
        if  f(a)*f(p) > 0
            a = p
        end
        
        if f(b)*f(p) > 0
            b = p
        end
        p= (a+b)/2
        push!(list,p)
    end
return    list #Se obtiene las raíces

end
```

```julia
biseccionraices(0,5,66) #Se prueba el código con las tres variables
```

```julia
list1=biseccionraices(0,50,44);#Listas de las raíces que se van a obtener
list2=biseccionraices(0,100,50);
list3=biseccionraices(0,120,66);
list4=biseccionraices(0,20,70);
```

```julia
plot(real(list1))#Se grafican las listas
plot(real(list2))
plot(real(list3))
plot(real(list4))
show()
```
