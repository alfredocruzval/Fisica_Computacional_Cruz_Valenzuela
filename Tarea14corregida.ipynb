Ejercicio 1: Resolver con Runge-Kutta orden 4 el sistema: \begin{align} \dot x &= y-y^3\ \dot y &= -x -y^2 \end{align}

    Grafiquen muchas trayectorias en el espacio de fase, utilizando condiciones iniciales aleatorias.

```julia
function RK4(f,list,x0)
     x = x0
     h = list[2]-list[1]
     listx=[]
     push!(listx,x)
     for i in 2:length(list)
        t = i*h
        k1 = f(x,t)
        k2 = f(x+(h/2)*k1,t+(h/2))
        k3 = f(x+(h/2)*k2, t+(h/2))
        k4 = f(x+h*k3, t+h)
        x = x+(h/6)*(k1+2*k2+2*k3+k4)
        push!(listx,x) 
     end
     return listx
end
```

```julia
using PyPlot

list6= 0:0.01:10

#La función que queremos resolver.
a(x,t)=[x[2]-x[2]^3,-x[1]-x[2]^2]

for i in 1:600
    ω = .01*rand(-500:500,2,1)
    solsrk=RK4(a,list6,ω);
    alistrk1=map(x->x[1],solsrk)
    alistrk2=map(x->x[2],solsrk)
    plot(alistrk1,alistrk2);
    xlim(-5,5)
    ylim(-5,5)
    grid()
end
```

Calculen el Jacobiano de $\vec F=\left(y-y^3,-x-y^2 \right)^{\text{T}}$ y evaluenlo en $\vec 0$, el cual es un punto fijo. Calculen $\Delta$ y $\tau$ e indiquen que tipo de punto fijo es el $\vec 0$.

Sea

$$\vec F=\left(\begin{array}{cc} y-y^3 \\ -x-y^2 \end{array} \right) = \vec 0$$

Se deben obtener los otros puntos fijos, esto se hace igualando los renglones a 0.

$$y-y^3 = 0$$$$-x-y^2 = 0$$

A partir de la primera ecuación: $$y(1-y^2) = 0$$

Ya que sabemos que los puntos críticos en $y$ son: $$y = 0,-1,1$$ y que en x son: $$y = 0,-1$$

Para el jacobiano se deben utilizar los puntos (-1,1) y en (-1,-1)

Ahora alculamos el jacobiano, entonces

$$J=\left(\begin{array}{cc} \frac{d f_1}{d x} ; \frac{d f_1}{d y} \\ \frac{d f_2}{d x} ; \frac{d f_2}{d y} \end{array} \right)$$

Evaluando el jacobiano en cero, se obtiene:
$$J = \left(\begin{array}{cc} 0 ; 1 \\ -1 ; 0 \end{array} \right)$$$$\tau=tr(J)$$$$\tau= 1$$$$\Delta= Det(J)$$$$\Delta= 1$$
